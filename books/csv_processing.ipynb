## data Null check

```python
df_user0.drop(index=df_user0[df_user0['Category'] == '9'].index, inplace=True) #remove 9 in category

df_user0.drop(index=df_user0[df_user0['rating'] == 0].index, inplace=True) #remove 0 in rating
```

```python
print("Null check")
print("============================")
print("user_id: ", df_user0.user_id.isnull().sum())
print("isbn: ", df_user0.isbn.isnull().sum())
print("book_title: ", df_user0.book_title.isnull().sum())
print("book_author: ", df_user0.book_author.isnull().sum())
print("year_of_publication: ", df_user0.year_of_publication.isnull().sum())
print("book_publisher: ", df_user0.book_publisher.isnull().sum())
print("Summary: ", df_user0.Summary.isnull().sum())
print("Category: ", df_user0.Category.isnull().sum())
```
